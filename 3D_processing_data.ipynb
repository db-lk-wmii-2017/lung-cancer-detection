```python
import os
import SimpleITK as sitk
import numpy as np
import pandas as pd 
import cv2
import math 

from os import walk
import matplotlib.pylab as plt
```

```python
images_path = '../data/'
des_path = '../completed/'
```

```python
IMG_SIZE = 64
PLT_HEIGHT = 4
PLT_WIDTH = 8

DATA_IMAGES = 1
DATA_DEPTH = 32
```

```python
def load_image(filename):
    image = sitk.ReadImage(filename)
    ct_scan = sitk.GetArrayFromImage(image)

    return ct_scan
```

```python
images = []
for (dirpath, dirnames, files) in walk(images_path):
    for file in files:
         if file.endswith(".mhd"):
            images.append((dirpath+file, os.path.splitext(file)[0]))
len(images)
```

```python
def chunks(l,n):
    start = 0
    for i in np.arange(0, len(l)-1, n):
        end = math.floor(i+n)
        yield l[start:end]
        start = end
```

```python
def mean(l):
    return sum(l)/len(l)
```

```python
def save_slices(file,slices):
    np.save(des_path+file, slices)
```

```python
def normalize_image(image):
    """From SITK - converting hounds units to grayscale units"""
    maxHU = 400.0
    minHU = -1000.0
    
    image = (image - minHU) / (maxHU - minHU)
    image[image > 1] = 1.0
    image[image < 0] = 0.0
    
    return (image * 255).astype(np.uint8);
```

```python
def load_slices(file):
    return np.load(des_path+file+".npy")
```

```python
def draw_slices(slices):
    fig = plt.figure(figsize=(50, 50))
    
    for i,testSlice in enumerate(slices):
        subplot = fig.add_subplot(PLT_HEIGHT,PLT_WIDTH,i+1)
        subplot.imshow(testSlice)
    plt.show()
```

```python
for index,image in enumerate(images, start=1):
    new_slices = []
    slices = load_image(image[0])
    
    chuck_size = len(slices) / DATA_DEPTH
    
    for slice_chunk in chunks(slices,chuck_size):
        slice_chunk = list(map(mean,zip(*slice_chunk)))
        new_slices.append(np.asarray(slice_chunk))
        
    new_slices = [cv2.resize(image,(IMG_SIZE,IMG_SIZE)) for image in new_slices]
    new_slices = [normalize_image(newslice) for newslice in new_slices]   
    
    print(str(index)+'/'+str(len(images)))
    #draw_slices(new_slices)
    save_slices(image[1],new_slices)
```
